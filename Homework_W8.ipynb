***Greedy Algorithm***  

````python
shifts = 24  
all_guards = 6 #total available guards
working_guards = 0 # we start at zero working guards
total_cost = 0 # zero starting cost
security_time =0 # total security time
final_guard_shifts = []

for guard in range(all_guards):
    guard_cost = 0 # guard cost before process
    hourly_cost = 15 
    use_avg_hourly =9999 # set arbitraly high start point for average hourly wage.
    overtime_cost = 5 
    overtime_threahold  = 8 # threshold for overtime
    guard_shift = 0 # set the beggining shift lenght for each guard
    
    for shift in range(1,shifts+1):
        
        if shift >  overtime_threahold: # adjust to account for overtime
            avg_hourly = (hourly_cost*shift + overtime_cost*(shift-overtime_threahold))/shift  

        else:
            avg_hourly =(hourly_cost*shift)/shift

        if avg_hourly <= use_avg_hourly: # readjust shift lenght per guard as long as averge hourly is less or equal to

            use_avg_hourly = avg_hourly
            guard_shift = shift
            guard_cost = avg_hourly*guard_shift
 
    # finalize the metrics
    final_guard_shifts.append(guard_shift)
    all_guards -=1 
    working_guards +=1     
    security_time += guard_shift
    total_cost += guard_cost
    
    if security_time >=24: #exit the loop when we reach 24 hours
        break
        
print("Total hours: ", security_time)
print("Total working guards: ", working_guards)
print("Total Cost: ", total_cost)
print("Remaining guards: " , all_guards)
print("Guard Shifts: " , final_guard_shifts)
  
        
````

***Explain your algorithm in detail.  How is it greedy?***

This greedy algorithm maximized the hours worked per guard at the lowest average hourly rate. We have 6 guards available that each can work from 0-24 hours shifts. There are many ways this can be solved. For example, we can try to determine the number of guards we need and the shift length for each guard that add up to 24 hours. This approach then assumes that each guard will work equal hours. 

I have taken a little different approach. In my greedy algorithm, we iterate over each guard and each shift and then minimize the average hourly cost. For example, we start with guard 1 and estimate the maximum hours to work ***given the average minimum wage***:

```python 
    if shift >  overtime_threahold: # adjust to account for overtime
        avg_hourly = (hourly_cost*shift + overtime_cost*(shift-overtime_threahold))/shift
    else:
        avg_hourly =(hourly_cost*shift)/shift
```

and we update the shift length for this guard (we intuitively know this will be an 8-hour shift): 
``` python
    if avg_hourly <= use_avg_hourly: # readjust shift lenght per guard as long as averge hourly is less or equal to
        use_avg_hourly = avg_hourly
        guard_shift = shift
        guard_cost = avg_hourly*guard_shift
```

We iterate over each shift and then we early stop the algorithm as soon as the total hours become 24 hours:
```python
    if security_time >=24: #exit the loop when we reach 24 hours
        break
```

Because we minimizing the function locally this then becomes a greedy algorithm by definition as we seek to minimize cost globally. After we minimize the cost for the first guard we move to the next guard and so on, until we reach 24 hours in total shifts. In this case, we know then we will employ 3 guards for 8 hours shift each.

Big O of this approach is O(G x S) - G=Guards, S=Shifts.In essence, in the worst case, we iterate over all guards and all shifts making this GxS time complexity.
